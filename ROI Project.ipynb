```python
class Calculator():
    def __init__(self,incomes,expenses,cashflows,rois):
        self.incomes = incomes
        self.expenses = expenses
        self.cashflows = cashflows
        self.rois = rois
    
    
    def income(self):
        rental_income = int(input("Please enter your total rental income: "))
        storage_income = int(input("Enter your total income from storage units: "))
        laundry_income = int(input("Enter your total income from laundry machines: "))
        miscellaneous_income = int(input("Enter your total miscellaneous income: "))
        income_total = rental_income + storage_income + laundry_income + miscellaneous_income
        print(f'Your total monthly income comes out to be : ${income_total}')
        self.incomes = income_total
    
    
    def expense(self):
        expense_tax = int(input(" Please enter total amount paid for taxes: "))
        expense_insurance = int(input("Enter your total monthly insurance payments: "))
        print("Will you be responsible for any of the utilities such as electic, water, sewer, or gas? :")
        expense_utilities = input("If you're paying ANY of these, type 'yes'. If the renter will be paying ALL the expenses type 'no'.  ")

        if expense_utilities.lower() == "yes":
            expense_electric = int(input("Please enter monthly total of electric bill payments: "))
            expense_water =int(input("Enter monthly total of water bill payments: "))
            expense_sewer = int(input("Enter total monthly sewage costs: "))
            expense_gas = int(input("Enter total monthly gas payments: "))
            total_utilities = expense_electric + expense_water + expense_sewer + expense_gas
            print(f"\nYour total monthly utility expenses comes out to be: ${total_utilities}")
        else:
            total_utilities = 0
            print("Okay, You don't have any utility expenses. Now lets look at your expenses. ")
        expense_hoa = int(input(" Please enter your monthly HOA: "))
        expense_lawn_snow = int(input("Enter total monthly expense for lawn care and/or snow removal:  "))
        expense_vacancy = int(input('Enter your total amount put aside in case of vacancy '))
        expense_repairs = int(input("Enter your monthly expenses on repairs: "))
        expense_capex = int(input("Enter your total monthly costs for capital expenditures: "))
        expense_propertymanagement = int(input("Enter your monthly property management fees: "))
        expense_mortgage = int(input("Enter your monthly mortgage payment: "))
        expense_total = expense_tax + expense_insurance + total_utilities + expense_hoa + expense_lawn_snow + expense_vacancy + expense_repairs + expense_capex + expense_propertymanagement + expense_mortgage
        self.expenses = expense_total
        print(f'Your total monthly expenses comes out to be : ${expense_total}')
   

    def cashflow(self):
        self.cashflows = self.incomes - self.expenses
        total_cashflow = self.cashflows
        print(f"Your Total monthly cashflow comes out to be : ${total_cashflow}")
    
    
    def roi(self):
        roi_downpayment = int(input("Enter your down payment: "))
        roi_closingcost = int(input("Enter your closing cost: "))
        roi_rehabbudget = int(input("Enter your total rehab budget: "))
        roi_miscellaneous = int(input("Enter your miscellaneous expeneses: "))
        total_investment = roi_downpayment + roi_closingcost + roi_rehabbudget + roi_miscellaneous
        print(f"Your investment total comes out to be : ${total_investment}")
       
       
        annual_total_flow = self.cashflows * 12
        
        
        self.rois = (annual_total_flow/total_investment)*100
        rois = self.rois
        print(f"\n\n\nYour total cash on cash roi comes out to be : {rois}%")

roiCalculator = Calculator([],[],[],[])


def run():
    while True:
        
        start = input("If you would like to begin, please type 'begin' or type 'quit'. ")
        if start.lower() == 'quit':
            print("GoodBye!")
            break
            
        if start.lower() == 'begin':
            print("First, we wil go over what your total monthly rental income. ")
            roiCalculator.income()
            print("\nSecond, we will go over your total monthly expenses. ")
            roiCalculator.expense()
            print("\nThen your monthly cash flow.")
            roiCalculator.cashflow()
            print("\n\nAnd finally, we will calculate the roi. ")
            roiCalculator.roi()

run()
```
